```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
media_data.head()
```

```python
media_data['Current Status'].unique()
```

```python
media_data.columns
```

```python
media_data.drop(['City', 'Location (City Airport Code)','State', 'Zone','Latitude', 'Longitude','WhatsApp users','Facebook users','Instagram users'], axis=1)
```

```python
features = ['Do you own multiple profiles on Instagram?','Gender','Highest Education','Phone OS']
```

```python
new_media_data = pd.get_dummies(media_data, columns = features)
```

```python
new_media_data.drop(['City', 'Location (City Airport Code)',
       'State', 'Zone', 'Latitude', 'Longitude', 'WhatsApp users', 'Facebook users',
       'Instagram users'], axis=1, inplace = True)
```

```python
new_media_data
```

```python
new_media_data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
new_media_data['Encoder_Current_Status']=label_encoder.fit_transform(new_media_data['Current Status'])
```

```python
new_media_data.head(10)
```

```python
new_media_data.drop('Current Status',axis=1,inplace=True)
```

```python
new_media_data.head()
```

```python
x = new_media_data.drop("Encoder_Current_Status", axis=1).values
y =new_media_data["Encoder_Current_Status"]
y = y.astype(int)
print(x.shape)
print(y.shape)
```

```python
dt = DecisionTreeClassifier(random_state=15, criterion = 'entropy', max_depth = 10)
dt.fit(x,y)
```

```python
fi_col=[]
fi = []

for i, column in enumerate(new_media_data.drop("Encoder_Current_Status", axis=1)):
    print("The feature importance for {} is : {}".format(column,dt.feature_importances_[i]))
    fi_col.append(column)
    fi.append(dt.feature_importances_[i])
```

```python
fi_col
fi 
fi_df=zip(fi_col,fi)
fi_df = pd.DataFrame(fi_df, columns=["Features","Features_importance"])
```

```python
fi_df=fi_df.sort_values("Features_importance", ascending=False).reset_index()
```

```python
columns_to_keep=fi_df["Features"][0:21]
```

```python
columns_to_keep
```

```python
fi_df
```

```python
x = new_media_data[columns_to_keep].values
y = new_media_data["Encoder_Current_Status"]
y = y.astype(int)
print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
dt_clf = DecisionTreeClassifier(random_state=15, criterion = 'entropy', max_depth = 5)
```

```python
dt_clf.fit(x_train,y_train)
```

```python
y_pred=dt_clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, r2_score, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
r2_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True)
```
